```python
#coding: utf-8 

import pandas as pd
import pickle
import networkx as nx
import matplotlib.pyplot as plt
import pickle
```

```python
df1 = pd.read_pickle("1970metadata2009.pkl")
df1.to_csv("df_author.csv")
df = pd.read_csv('df_author.csv', encoding = 'utf8')
df
```

```python
G = nx.from_pandas_edgelist(df, 'DOI','Venue',)
```

```python
df_DOI = pd.DataFrame(dict(
    DEGREE = dict(G.degree),
    DEGREE_CENTRALITY = nx.degree_centrality(G),
    EIGENVECTOR = nx.eigenvector_centrality(G,tol=1e-03),))
```

```python
df_DOI
```

```python
TV = nx.from_pandas_edgelist(df, 'Title','Venue')
```

```python
TV_author = pd.DataFrame(dict(
    DEGREE = dict(TV.degree),
    DEGREE_CENTRALITY = nx.degree_centrality(TV),
    EIGENVECTOR = nx.eigenvector_centrality(TV,tol=1e-03),))
```

```python
TV_author
```

```python
result = df_DOI.merge(df_DOI, TV_author, how="right", on=["DEGREE", "DEGREE"])
display(result)
```
